# ENCODING DATA

```python
from sklearn import preprocessing
```

```python
label_enc=preprocessing.LabelEncoder()
```

```python
s=df.copy()
```

```python
for col in s:
    s[col]=label_enc.fit_transform(s[col].values)
```

```python
s
```

# SMOTE of data set

```python

from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(s.iloc[:,0:-1], s['Income'])

print('Original dataset shape', len(s))
print('Resampled dataset shape', len(y_smote))
```

# Spliting The Data for Testing And Trainng

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_smote, y_smote,test_size=0.25,random_state=42)
```

```python
x_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logicreg=LogisticRegression()
```

```python
logicreg
```

#### In a Jupyter environment, please rerun this cell to show the HTML representation or trust the notebook. On GitHub, the HTML representation is unable to render, please try loading this page with nbviewer.org.

```python
logicreg.fit(x_train,y_train)
```

```python
y_predict=logicreg.predict(x_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
acc=accuracy_score(y_test,y_predict)
```

```python
acc
```

```python
con=confusion_matrix(y_test,y_predict)
```

```python
con
```

```python
f1=f1_score(y_test,y_predict)
```

```python
f1
```

```python
pre=precision_score(y_test,y_predict)
```

```python
pre
```

```python
recall=recall_score(y_test,y_predict)
```

```python
recall
```

```python
print("ACCURACY:",acc)
print("PRECISION:",pre)
print("RECALL:",recall)
print("F1 SCORE:",f1)
```
